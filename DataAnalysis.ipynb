# Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
```

# Analysis

```python
file_name = 'olympics.csv'
df = pd.read_csv(file_name, skiprows = 4)
df.head()
```

# In which events did Jesse Owens win a medal?

```python
jo = df[df.Athlete == 'OWENS, Jesse']
jo['Event'].value_counts()
```

# Which country has won the most men's gold medals in singles badminton over the years? Alphabetically sorted by the player's names

```python
gbm = df[(df.Medal == 'Gold') & (df.Gender == 'Men') & (df.Sport == 'Badminton')]
gbm.sort_values(by='Athlete')
```

# Which three countries have won the most medals in recent years (from 1984 to 2008)?

```python
recent = df[df.Edition >= 1984]
recent.NOC.value_counts().head(3)
```

# Display the male gold medal winners for the 100m Track & Field sprint event over the years. List the results starting with the most recent. Show the Olympic City, Edition, Athlete and the country they represent

```python
gmh = df[(df.Gender == 'Men') & (df.Medal == 'Gold') & (df.Event == '100m')]
gmh.sort_values(by='Edition',ascending = False)[['City','Edition','Athlete','NOC']]
```

## Which country did not win a medal in the 2008 Olympics? How many countries were there?

```python
# store a data frame to display all of the countries who participated in the 2008 edition
# of the olympics

lo = df[df.Edition == 2008]
lo.head()
```

```python
# using another df, that countains all of the countries who participated in the olympics
# create another series with the value counts of the numbers of medals of every NOC of 2008
# edition and insert in in the participants dataframe, so that it can display the NaN values
# if the NOC didn't win a medal

noc = pd.read_csv('Summer Olympic medallists 1896 to 2008 - IOC COUNTRY CODES.csv')
noc.set_index('Int Olympic Committee code',inplace = True)
medal_2008 = lo.NOC.value_counts()
noc['medal_2008'] = medal_2008
noc[noc.medal_2008.isnull()]
```

## Display the total numbers of medals awarded at each of the Olympic games throughout history

```python
df.groupby(['Edition','Event']).agg({'Medal':['count']})
```

## Display the total number of medals won for each country over the history of the Olympics. For each country, include the year of the first and most recent Olympic medal wins

```python
df.groupby('NOC').agg({'Medal' : ['count'], 'Edition' : ['min','max']})
```

## Plot the number of gold medals won by the US male and female athletes throughout the history of the Olympics

```python
gold = df[(df.Medal == 'Gold') & (df.NOC == 'USA')]
gold.groupby(['Edition','Gender']).size().unstack('Gender',fill_value=0).plot()
```

## Plot the 5 athletes who have won the most gold medals over the history of the Olympics. When there is a tie, consider the number of silver medals, then bronze medals

```python
g = df.groupby(['Athlete','Medal']).size().unstack('Medal',fill_value = 0)
g.sort_values(['Gold','Silver','Bronze'],ascending = False)[['Gold','Silver','Bronze']].head(5).plot(kind='bar')
```

# In every Olympics, which US athlete has won the most total number of medals? Include the athlete's discipline

```python
gy = df[df.NOC == 'USA']
gy = gy.groupby(['Edition','Athlete','Medal']).size().unstack('Medal',fill_value = 0)
gy['Total'] = gy['Gold']+gy['Silver']+gy['Bronze']
gy.reset_index(inplace=True)
tu = [group.sort_values('Total',ascending = False)[:1] for year, group in gy.groupby('Edition')]
top = pd.DataFrame()
for i in tu:
    top = top.append(i)
top.sort_values('Total',ascending = False)[:1]
```
